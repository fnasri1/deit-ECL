```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
f = open('deit_tiny.txt',)
data = json.load(f)
```

```python
data
```

```python
train_lr = []
train_loss = []
test_loss = []
test_acc1 = []
test_acc5 = [] 
epoch = []
for ele in data:
    train_lr.append(ele['train_lr'])
    train_loss.append(ele['train_loss'])
    test_loss.append(ele['test_loss'])
    test_acc1.append(ele['test_acc1'])
    test_acc5.append(ele['test_acc5'])
    epoch.append(ele['epoch'])
```

```python
df = pd.DataFrame()
df['train_lr'] = train_lr
df['train_loss'] = train_loss
df['test_loss'] = test_loss
df['test_acc1'] = test_acc1
df['test_acc5'] = test_acc5
df['epoch'] = epoch
```

```python
plt.plot(df['epoch'], df['train_lr'])
```

```python
plt.plot(df['epoch'], df['train_loss'])
```

```python
plt.plot(df['epoch'], df['test_loss'])
```

```python
plt.plot(df['epoch'], df['test_acc1'])
```

```python
plt.plot(df['epoch'], df['test_acc5'])
```
